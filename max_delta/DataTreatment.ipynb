```python
import pandas as pd
import datetime as dt
```

```python
text = open('weight_history.txt', 'r').read()
```

```python
listofvalues = text.split('\n')
```

```python
# Corrects ':' date separator (instead of patterned '/') to '/' and lists entries
l = list()


for item in listofvalues:
    newitem = list(tuple(item.split(':')))
    # print(type(newitem))
    # print(newitem)
    if len(newitem) > 2:
        newitem1=newitem[0]
        newitem2=newitem[1]
#         print( newitem1+'/'+newitem2)
        
        newitem[0]=(newitem1+'/'+newitem2)
        newitem.pop(1)
    l.append(newitem)

print(l)
```

```python
l
```

```python
# Cleans empty entries (Week separator spaces in original dataset)
for entry in l:
    if len(entry)<2:
        l.remove(entry)
```

```python
# Removes header entry
l.remove(l[0])
```

```python
l
```

```python
# Converts dates to datetime types
for entry in l:

    datestr = str(entry[0])
    print(datestr)
    
    if len(datestr)==10:  # Case date formatted like 'dd/MM/YYYY'
#         print(datestr)
        entry[0] = dt.datetime.strptime(datestr, '%d/%m/%Y')
    elif len(datestr)==8: # Case date formatted like 'dd/MM/YY'
#         print(datestr)
        entry[0] = dt.datetime.strptime(datestr[:6] + '20' + datestr[6:], '%d/%m/%Y')
    elif len(datestr)==5: # Case date formatted like 'dd/MM'
#         print(datestr)
#         print(datestr[3:])
        premisseYear = datestr+'/2018' if datestr[3:]=='12' else datestr+'/2019'
#         print(premisseYear)
        entry[0] = dt.datetime.strptime(premisseYear, '%d/%m/%Y')
```

```python
l
# Datetimes typed, wheight measurements stringged
```

```python
# Converting wheight measurements to float
for entry in l:
    wheightString = entry[1]
    entry[1] = float(wheightString.replace('kg', '').replace(',','.'))
```

```python
l
```

```python
# Correcting corner case of wheight measurement formatted like 'XXXX' without ','
for entry in l:
    if entry[1] > 1000:
        entry[1] = entry[1]/10
```

```python
l
```

```python
# Self Challenge: What is the biggest wheight loss?

lastWheight = df.wheight.max()
distance = 0

for entry in df.iterrows():
    
#     print(entry)
    l = dict(entry[1])
#     print(l)
    
    NewPotencialDistance = abs(l['wheight'] - lastWheight)
#     print(NewPotencialDistance)
    
    if NewPotencialDistance > distance:
        distance = NewPotencialDistance
        date = l['date']
    
    lastWheight = l['wheight']

print('A maior perda de peso foi de {0}kg no dia {1}'.format(str(distance).replace('.',',',), date))
```
